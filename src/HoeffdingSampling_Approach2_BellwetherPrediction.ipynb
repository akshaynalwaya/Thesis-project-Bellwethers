```python
import os, random
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix

random.seed(10)
```

```python
# os.getcwd()
data_path = os.getcwd()+"/defects/src/data/Jureczko/collated_data/"
# os.chdir(os.getcwd() + data_path)
```

```python
projList = os.listdir(data_path)
print("Files:",projList)
projs = [p.split('_')[0] for p in projList]
print("List of projects:",projs)
```

```python
df_n = pd.read_csv(os.getcwd()+"/defects/src/data/Jureczko/"+"/samplingResults_n.csv")
fractions = []

for i in range(10):
    frac = 0
    for j in range(10):
        frac = max(frac,df_n.iloc[i][j+1])
    fractions.append(frac/100)
len(fractions)
# df_n
```

```python
fractions
```

```python
import time

def getRuntimes(randomSeed):
    random.seed(randomSeed)
    tbl = [[0]*len(projs) for p in projs]
    count=-1
    startTime = time.time()
    for i in range(len(projs)):
        print("\nIteration ",i)
        trainData = pd.read_csv(data_path + projList[i])
        trainData.sample(frac=fractions[i])
        X_train = trainData.loc[:,trainData.columns!='$<bug']
        y_train = trainData.loc[:,trainData.columns=='$<bug']
        clf = RandomForestClassifier(n_estimators=1000, n_jobs=1)
        clf.fit(X_train, y_train.values.ravel())
                
        for j in range(len(projs)):
            count+=1
            if (i != j):
                #print("Here: ",float(df_n.iloc[i][j+1]/100))
                
                testData = pd.read_csv(data_path + projList[j])
                X_test = testData.loc[:, testData.columns!='$<bug']
                y_test = testData.loc[:, testData.columns=='$<bug']
                y_pred = clf.predict(X_test)
                cm = confusion_matrix(y_test, y_pred)
                recall = cm[1][1]/(cm[1][1] + cm[1][0])
                pf = cm[0][1]/(cm[0][1] + cm[0][0])

                g = 2/((1/recall) + (1/(1-pf)))

                #print("Model trained on",projs[i],", testing on",projs[j])
                #acc = metrics.accuracy_score(y_test, y_pred)
                #print("Project: {}, Accuracy: {}, Precision: {}".format(projs[j],acc,
                #                                                       metrics.precision_score(y_test, y_pred)))
                tbl[i][j]=g
                #print("Test Project: {}, G-Score {}".format(projs[j], g))
    #print(tbl)
    return (time.time() - startTime),tbl
```

```python
runtimes = []
for i in range(30):
    print(i)
    runtime,tbl = getRuntimes(i)
    runtimes.append(runtime)
```
